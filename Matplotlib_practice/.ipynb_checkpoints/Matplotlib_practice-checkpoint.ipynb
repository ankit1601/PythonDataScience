```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
# that enable us to see output inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#FUNCTIONAL METHOD
plt.plot(x,y) #plt.show() print plot
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
#Mulriplot on same canvas
plt.subplot(1,2,1) #num of row , num of col and plot number args
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#Object Oriented Method
fig = plt.figure() #it will generate figure object
#consider that as imaginary blank canvas
axes = fig.add_axes([0.1,0.1,0.8,0.8])#list have left bottom width and height

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('y label')
axes.set_title('Title')
```

```python
axes
```

```python
fig = plt.figure() #create figure obj
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Bigger plot')
axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
#SUBPLOT USING OO PROGRAMMING
fig,axes = plt.subplots(nrows=1,ncols=2)
# for current_ax in axes:
#     current_ax.plot(x,y)

axes[0].plot(x,y)
axes[0].set_title("First plot")

axes[1].plot(y,x)
axes[1].set_title('Second plot')
```

```python
axes
```

<h1>Figure Size and DPI</h1>

```python
fig = plt.figure(figsize=(3,2))#,dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
#SAVE A FIGURE
fig.savefig('my_picture.png',dpi=200)
```

```python
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('Title')
ax.set_xlabel('X')
ax.set_ylabel('Y')
```

```python
#LEGENDS

fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cubed')
ax.set_title('Title')
ax.set_xlabel('X')
ax.set_ylabel('Y')
#ax.legend(loc=(0.1,0.1))
ax.legend(loc=0)
```

```python
#SETTING COLORS LINE WIDTHS , LINE TYPES

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y , color='green',lw=3,ls='--',marker='o',markersize=10,markerfacecolor='yellow',markeredgewidth=2,markeredgecolor='green')#,linestyle=':'),linewidth=3,alpha=0.5) #you can put rgb hex code as well
```

```python
ax.plot(x,y , color='green',linewidth=3,alpha=0.5)
```

```python
#SETTING COLORS LINE WIDTHS , LINE TYPES

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y , color='green',lw=2,ls='--') #you can put rgb hex code as well
ax.set_xlim(0,1)
ax.set_ylim(0,1)
```
